*String data converted to numerical data using encoding*

   get_dummy

```python
dummy=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

*Combine 2 dataframe*

```python
dfe=pd.concat([df,dummy],axis=1)
dfe
```

*To drop unwanted columns*

```python
df1=dfe.drop(['Name','Location','Fuel_Type','Unnamed: 0','Transmission','Owner_Type','New_Price','Fuel_Type_Electric'],axis=1)
df1
```

*To replace string values*

```python
df1['Mileage']=df1['Mileage'].str.replace('km/kg','')
df1
```

```python
df1['Mileage']=df1['Mileage'].str.replace('kmpl','')
df1
```

```python
df1['Engine']=dfe['Engine'].str.replace('CC','')
df1
```

```python
df1['Power']=df1['Power'].str.replace('bhp','')
df1
```

```python
#Milegae some values are type as null.that null was replace by 0
df1['Mileage']=df1['Mileage'].str.replace('null','0')
df1
```

```python
df1['Engine']=df1['Engine'].str.replace('null','0')
df1
```

```python
df1['Power']=df1['Power'].str.replace('null','0')
df1
```

```python
df1.isna().sum()
```

```python
df1.dtypes
```

```python
#to convert milegae,power,engine into float
df1['Mileage']=df1['Mileage'].astype(float)
df1['Engine']=df1['Engine'].astype(float)
df1['Power']=df1['Power'].astype(float)
```

```python
df1.dtypes
```

```python
#mileage engine power is now 0 that conveted into float that replace into not a numbr(NaN)
df1.loc[df1.Engine==0,'Engine']=np.NaN
df1.loc[df1.Power==0,'Power']=np.NaN
df1.loc[df1.Mileage==0,'Mileage']=np.NaN
```

```python
df1.isna().sum()
#to increse the missing value becoz null valu act as missing value
```

```python
#seat mode
#mean
df1['Engine']=df1['Engine'].fillna(df1['Engine'].mean())
df1['Power']=df1['Power'].fillna(df1['Power'].mean())
df1['Mileage']=df1['Mileage'].fillna(df1['Mileage'].mean())
df1['Seats']=df1['Seats'].fillna(df1['Seats'].mode() [0])
```

```python
df1.isna().sum()
```

```python
#To seperate x
x=df1.drop(['Price'],axis=1)
x
```

```python
#to seprate y
y=df1['Price']
y
```
